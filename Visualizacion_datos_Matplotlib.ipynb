```python
import pandas as pd
```

```python
df=pd.read_csv('base_tiempo.csv')
```

```python
df.head()
```

```python
#convertimos fecha en formato data-time
df['date']= pd.to_datetime(df['fecha'])
df.info()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#ver la temperatura en el tiempo (date)
plt.plot(df['date'], df['temperatura']) #eje x - y
```

```python
#cambiar configuracion de cuadro
plt.figure(figsize=(15,8))
plt.plot(df['date'], df['temperatura'])
plt.title('temperatura en el tiempo');
```

```python
#CAMBIAR a EJES
fig= plt.figure(figsize=(15,8))
eje=fig.add_axes([0, 0, 1, 1])
eje.plot(df['date'], df['temperatura'], color='green') #colocar nombre a los ejes y color
eje.set_title('Temperatura en el tiempo', fontsize=25)
eje.set_ylabel('temperatura', fontsize=20)#colocar nombre al eje y cambiar de tamaño
eje.set_xlabel('date', fontsize=20)
eje.legend(['temperatura'], loc='lower right', fontsize=15)
```

personalizar visualizaciones

```python
fig= plt.figure(figsize=(15,8))
eje=fig.add_axes([0, 0, 1, 1])
eje.plot(df['date'], df['temperatura'], color='green', lw=0.4) #cambiar grosor de lineas(lw)
eje.set_title('Temperatura en el tiempo', fontsize=25)
eje.set_ylabel('temperatura', fontsize=20)#colocar nombre al eje y cambiar de tamaño
eje.set_xlabel('date', fontsize=20)
eje.legend(['temperatura'], loc='lower right', fontsize=15)
```

```python
fig= plt.figure(figsize=(15,8))
eje=fig.add_axes([0, 0, 1, 1])
eje.plot(df['date'], df['temperatura'], color='green', lw=0.4, ls='dotted') #ls cambia tipo de linea
eje.set_title('Temperatura en el tiempo', fontsize=25)
eje.set_ylabel('temperatura', fontsize=20)#colocar nombre al eje y cambiar de tamaño
eje.set_xlabel('date', fontsize=20)
eje.legend(['temperatura'], loc='lower right', fontsize=15)
```

```python
import datetime
fig= plt.figure(figsize=(15,8))
eje=fig.add_axes([0, 0, 1, 1])
eje.plot(df['date'], df['temperatura'], color='green',)
eje.set_xlim(datetime.datetime(2014,1,1), datetime.datetime(2015,1,1))#limitar datos en el tiempo
eje.set_title('Temperatura en el tiempo', fontsize=25)
eje.set_ylabel('temperatura', fontsize=20)
eje.set_xlabel('date', fontsize=20)
eje.legend(['temperatura'], loc='lower right', fontsize=15)
eje.grid(True) #cuadricula de limites
```

```python
#dos tipos de informacion en un mismo grafico
fig= plt.figure(figsize=(15,8))
eje=fig.add_axes([0, 0, 1, 1])
eje2=fig.add_axes([0.7, 0.65, 0.3, 0.3])
eje.grid(True)
eje.plot(df['date'], df['temperatura'], color='green',)
eje.set_xlim(datetime.datetime(2014,1,1), datetime.datetime(2015,1,1))
eje.set_ylabel('temperatura', fontsize=20)
eje.set_xlabel('date', fontsize=20)
eje.set_title('Temperatura en el 2014', fontsize=25, pad=20)
eje.legend(['temperatura'], loc='lower right', fontsize=15)

eje2.plot(df['date'], df['temperatura'], color='b',)
eje2.set_ylabel('temperatura', fontsize=10)
eje2.set_xlabel('date', fontsize=10)
eje2.set_title('Temperatura en el tiempo', fontsize=15)
eje2.legend(['temperatura'], loc='best', fontsize=8)
eje2.grid(True)
```

anotaciones y otros graficos

```python
fig= plt.figure(figsize=(15,8))
eje=fig.add_axes([0, 0, 1, 1])
eje.grid(True)
eje.plot(df['date'], df['temperatura'], color='green',)
eje.set_ylabel('temperatura', fontsize=20)
eje.set_xlabel('date', fontsize=20)
eje.set_title('Temperatura', fontsize=25, pad=20)
eje.legend(['temperatura'], loc='lower right', fontsize=15)

#determinar los puntos maximos y minimos
x1= df['date'] [df['temperatura'].idxmax()]
y1= max(df['temperatura'])
#colocar linea sobre el punto maximo y minimo
eje.axhline(max(df['temperatura']), color='k', linestyle='--')
eje.axhline(min(df['temperatura']), color='k', linestyle='--')
#realizamos una anotacion en el maximo y minimo
eje.annotate('maximo', xy=(x1,y1), fontsize=20)
```

```python
fig= plt.figure(figsize=(15,8))
eje=fig.add_axes([0, 0, 1, 1])
eje.grid(True)
eje.plot(df['date'], df['temperatura'], color='green',)
eje.set_ylabel('temperatura', fontsize=20)
eje.set_xlabel('date', fontsize=20)
eje.set_title('Temperatura', fontsize=25, pad=20)
eje.legend(['temperatura'], loc='lower right', fontsize=15)
#generar una flecha de marcacion
x1= df['date'] [df['temperatura'].idxmax()]
y1= max(df['temperatura'])
x2= df['date'] [df['temperatura'].idxmax() - 7000] #7000 equivalentes a hora
y2= max(df['temperatura']-5)
eje.annotate('maximo', xy=(x1,y1), fontsize=20, xytext=(x2,y2), arrowprops=dict(facecolor='k'))
x1= df['date'] [df['temperatura'].idxmin()]
y1= min(df['temperatura'])
x2= df['date'] [df['temperatura'].idxmin() - 7000]
y2= min(df['temperatura']+5)
eje.annotate('minimo', xy=(x1,y1), fontsize=20, xytext=(x2,y2), arrowprops=dict(facecolor='k'))

eje.axhline(max(df['temperatura']), color='k', linestyle='--')
eje.axhline(min(df['temperatura']), color='k', linestyle='--')

```

agrupamos datos por...

```python
temperatura_por_dia_de_semana= df.groupby('dia_de_semana')['temperatura'].mean()
nombre_dias= ['Domingo', 'Lunes', 'Martes', 'Miercoles', 'Jueves', 'Viernes', 'Sabado']
temperatura_por_dia_de_semana = temperatura_por_dia_de_semana[nombre_dias]
temperatura_por_dia_de_semana
```

```python
#elaboramos grafica
fig= plt.figure(figsize=(5,4))
eje= fig.add_axes([0,0,1,1])
indice= range(len(temperatura_por_dia_de_semana))

eje.bar(indice, temperatura_por_dia_de_semana, edgecolor='black')#edgecolor=colocar marco en barras
eje.set_title('Temperatura por dia de semana',fontsize=15, pad=10)
eje.set_xlabel('Dia de semana', fontsize=15)
eje.set_ylabel('Temperatura promedio', fontsize=10)
#cambiar valores de los ejes
eje.set_xticks(indice)
eje.set_xticklabels(nombre_dias);
```

```python
fig= plt.figure(figsize=(5,4))
eje= fig.add_axes([0,0,1,1])
#grafico de pizza
eje.pie(temperatura_por_dia_de_semana, labels=temperatura_por_dia_de_semana.index, autopct='%.1f%%')#autopct=establecer porcentaje ,decimales
eje.set_title('Temperatura por dia de semana', fontsize=15, pad=10);
```

#graficos de dispersion

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df_iris= pd.read_csv('iris.csv')
```

```python
df_iris.head()
```

```python
fig= plt.figure(figsize=(15,8))
ejes=fig.add_axes([0,0,1,1])
ejes.scatter(df_iris['SepalLengthCm'], df_iris['SepalWidthCm'])
ejes.set_title('Grafico de Dispersion', fontsize=25, pad=15)
ejes.set_xlabel('SepalLengthCm', fontsize=15)
ejes.set_ylabel('SepalWidthCm', fontsize=15)
ejes.tick_params(labelsize=15) #cambiar el tamaño de los tick
```

```python
#diferenciar cada clase del dataset en el grafico con colores y marcadores
fig= plt.figure(figsize=(15,8))
ejes=fig.add_axes([0,0,1,1])

colores={'Iris-setosa':'r', 'Iris-versicolor':'b', 'Iris-virginica':'g'}
marcadores={'Iris-setosa':'x', 'Iris-versicolor':'o', 'Iris-virginica':'v'}

for especie in df_iris['Species'].unique():
  tmp=df_iris[df_iris['Species']== especie]
  ejes.scatter(tmp['SepalLengthCm'], tmp['SepalWidthCm'], color=colores[especie], marker=marcadores[especie])

ejes.set_title('Grafico de Dispersion', fontsize=25, pad=15)
ejes.set_xlabel('SepalLengthCm', fontsize=15)
ejes.set_ylabel('SepalWidthCm', fontsize=15)
ejes.tick_params(labelsize=15)
ejes.legend(colores, fontsize=20);
```

boxplot

```python
fig= plt.figure(figsize=(8,5))
ejes=fig.add_axes([0,0,1,1])
ejes.boxplot(df_iris.drop('Species', axis=1).values, patch_artist=True)
ejes.set_title('Graficos de caja', fontsize=15, pad=10)
ejes.set_xticklabels(df_iris.drop('Species', axis=1).columns);
```

```python
#agregar color a cada caja
fig= plt.figure(figsize=(8,5))
ejes=fig.add_axes([0,0,1,1])

colores=['red', 'blue', 'orange', 'green']
cajas= ejes.boxplot(df_iris.drop('Species', axis=1).values, patch_artist=True)
#zip para juntar los valores que tienen igual cantidad, boxes=elegir especficamente las cajas del grafico
for caja,color in zip(cajas['boxes'], colores):
 caja.set(color=color)

ejes.set_title('Graficos de caja', fontsize=15, pad=10)
ejes.set_xticklabels(df_iris.drop('Species', axis=1).columns);
```

histograma

```python
fig= plt.figure(figsize=(5,4))
ejes=fig.add_axes([0,0,1,1])

ejes.hist(df_iris['PetalLengthCm'])
ejes.set_title('Histograma', fontsize=15, pad=10)
ejes.set_xlabel('PetalLengthCm', fontsize=15)
ejes.grid(True)
```

```python
fig= plt.figure(figsize=(5,4))
ejes=fig.add_axes([0,0,1,1])

ejes.hist(df_iris['PetalLengthCm'], bins=20, density=True) #bins para agregar mas columnas , densidad estandariza los datos en una escala de 0 a 1
ejes.set_title('Histograma', fontsize=15, pad=10)
ejes.set_xlabel('PetalLengthCm', fontsize=15)
ejes.grid(True)
```

personalizando histogramas

```python
fig= plt.figure(figsize=(5,4))
ejes=fig.add_axes([0,0,1,1])
#calcular media y desviacion estandar
mu, sigma = df_iris['PetalLengthCm'].mean(), df_iris['PetalLengthCm'].std()

ejes.hist(df_iris['PetalLengthCm'], bins=20)
ejes.set_title('Histograma', fontsize=15, pad=10)
ejes.set_xlabel('PetalLengthCm', fontsize=15)
ejes.grid(True)
ejes.annotate('$mu={0:.2f}$\n$sigma={1:.2f}$'.format(mu, sigma), xy=(4.5, 20), fontsize=20) #colocar formulas en formato texto
ejes.axvline(mu, color='k', linestyle='--')
ejes.annotate('promedio', xy=(mu-1.3, 28), fontsize=15)
```
